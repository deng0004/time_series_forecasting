```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import SimpleRNN, Dense, LSTM, GRU, Conv1D, MaxPooling1D, Flatten

# Load and preprocess your data
data = pd.read_csv('data4.csv')[:31276]
data['timestamp'] = pd.to_datetime(data['timestamp'], unit='s')
ts_data = data.set_index('timestamp')['MQ2_LPG']

# Normalize the data
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(np.array(ts_data).reshape(-1, 1))

# Split data into training and testing sets
train_size = int(0.8 * len(scaled_data))
train_data = scaled_data[:train_size]
test_data = scaled_data[train_size:]

# List of different window sizes to test
window_sizes = [5, 10, 15]

# Initialize dictionaries to store results
mse_results = {
    'LSTM': [],
    'Dense': [],
    'GRU': [],
    'CNN': [],
    'SimpleRNN': []
}

# Prepare data for the models
def create_dataset(dataset, time_steps=1):
    dataX, dataY = [], []
    for i in range(len(dataset) - time_steps):
        a = dataset[i:(i + time_steps), 0]
        dataX.append(a)
        dataY.append(dataset[i + time_steps, 0])
    return np.array(dataX), np.array(dataY)

# Iterate over different window sizes
for time_steps in window_sizes:
    X_train, y_train = create_dataset(train_data, time_steps)
    X_test, y_test = create_dataset(test_data, time_steps)

    # Reshape input to [samples, time_steps, features]
    X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
    X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

    # Train and evaluate models
    models = {
        'LSTM': Sequential([LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)),
                            LSTM(units=50, return_sequences=False),
                            Dense(units=1)]),
        'Dense': Sequential([Dense(units=50, activation='relu', input_dim=time_steps),
                             Dense(units=1)]),
        'GRU': Sequential([GRU(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)),
                           GRU(units=50, return_sequences=False),
                           Dense(units=1)]),
        'CNN': Sequential([Conv1D(filters=64, kernel_size=2, activation='relu', input_shape=(X_train.shape[1], 1)),
                           MaxPooling1D(pool_size=2),
                           Flatten(),
                           Dense(units=50, activation='relu'),
                           Dense(units=1)]),
        'SimpleRNN': Sequential([SimpleRNN(units=50, input_shape=(X_train.shape[1], 1)),
                                 Dense(units=1)])
    }

    for model_name, model in models.items():
        model.compile(optimizer='adam', loss='mean_squared_error')
        model.fit(X_train, y_train, epochs=50, batch_size=64, verbose=0)
        predictions = model.predict(X_test)
        mse = mean_squared_error(scaler.inverse_transform(y_test.reshape(-1, 1)),
                                 scaler.inverse_transform(predictions))
        mse_results[model_name].append(mse)

# Plot the mean squared error for different models
plt.figure(figsize=(10, 6))
for model_name, mse_values in mse_results.items():
    plt.plot(window_sizes, mse_values, marker='o', label=model_name)

plt.xlabel('Window Size')
plt.ylabel('Mean Squared Error')
plt.title('Mean Squared Error for Different Window Sizes and Models')
plt.legend()
plt.grid(True)
plt.show()

# Display MSE results in a table
mse_table = pd.DataFrame(mse_results, index=window_sizes)
print("Mean Squared Error Results:")
print(mse_table)
```

From the results above, we can see that the LSTM model is the worst model, because its has the highest different of Mean Squared Error (MSE). SimpleRNN model is the best model, because its has the lowest different of Mean Squared Error (MSE).
